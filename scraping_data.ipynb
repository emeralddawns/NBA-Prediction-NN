# Web Scrape NBA Game Data

```python
# Import dependencies
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.support.ui import Select
```

```python
# Create a Web Driver Object
driver = webdriver.Chrome(ChromeDriverManager().install())
```

## Pull Regular Season Data

```python
# Connect to the website
driver.get(url="https://www.nba.com/stats/teams/boxscores/?Season=2021-22&SeasonType=Regular%20Season")
```

```python
# Select the drop down element
select = Select(driver.find_element_by_class_name("stats-table-pagination__select"))
```

```python
# Select the "ALL" option, which is Index 0
select.select_by_index(0)
```

```python
# Read the table on the webpage
tbl = driver.find_element_by_tag_name("table").get_attribute('outerHTML')
```

```python
# Convert the table (list) into a DataFrame
result = pd.read_html(tbl)
```

```python
# Pull the table into a DataFrame
table_df = result[0]
table_df
```

```python
# Drop unnecessary columns
table_df = table_df.drop(['Match\xa0Up', 'Game\xa0Date', 'Season', 'W/L', 'FG%', '3P%', 'FT%'], axis=1)
```

```python
# Rename the columns to match the database schema
table_df = table_df.rename(columns={"Team":"team"})
table_df = table_df.rename(columns={"MIN":"min", "PTS":"pts", "FGM":"fgm"})
table_df = table_df.rename(columns={"FGA":"fga", "3PM":"threepm", "3PA":"threepa", "FTM":"ftm", "FTA":"fta"})
table_df = table_df.rename(columns={"OREB":"oreb", "DREB":"dreb", "REB":"reb", "AST":"ast", "STL":"stl"})
table_df = table_df.rename(columns={"BLK":"blk", "TOV":"tov", "PF":"pf", "+/-":"point_diff"})
```

```python
# Export the dataframe into a CSV file
table_df.to_csv("NBA_Data_Reg_Season_2022.csv")
```
