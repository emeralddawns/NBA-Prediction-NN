```python
# Import dependencies
from bs4 import BeautifulSoup as bs
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
import lxml.html as lh
```

```python
# URL of page to be scraped
url = 'https://www.nba.com/stats/teams/boxscores/?Season=2021-22&SeasonType=Regular%20Season'

# browser = init_browser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# launch browser
browser.visit(url)
html = browser.html
```

```python
# Create BeautifulSoup object, parse with 'html.parser'
soup = bs(html, 'html.parser')

# Extract feature image data
table = soup.find('table')
```

```python
# Display the results
print(table)
```

```python
# Create a DataFrame to store the data
df = pd.DataFrame(columns=['Team','Match Up','Game Date','W/L','MIN','PTS','FGM','FGA','FG%','3PM','3PA','3P%','FTM','FTA','FT%','OREB','DREB','REB','AST','STL','BLK','TOV','PF','+/-'])
df
```

```python
# Collect the data from the scrape 
for row in table.tbody.find_all('tr'):

    # Find all data for each column
    columns = row.find_all('td')

    if (columns != []):
        team = columns[0].text.strip()
        match_up = columns[1].text.strip()
        game_date = columns[2].text.strip()
        w_l = columns[4].text.strip()
        min = columns[5].text.strip()
        pts = columns[6].text.strip()
        fgm = columns[7].text.strip()
        fga = columns[8].text.strip()
        fgp = columns[9].text.strip()
        tpm = columns[10].text.strip()
        tpa = columns[11].text.strip()
        tpp = columns[12].text.strip()
        ftm = columns[13].text.strip()
        fta = columns[14].text.strip()
        ftp = columns[15].text.strip()
        oreb = columns[16].text.strip()
        dreb = columns[17].text.strip()
        reb = columns[18].text.strip()
        ast = columns[19].text.strip()
        stl = columns[20].text.strip()
        blk = columns[21].text.strip()
        tov = columns[22].text.strip()
        pf = columns[23].text.strip()
        pm = columns[24].text.strip()

        df = df.append({'Team':team, 'Match Up':match_up, 'Game Date':game_date, 'W/L':w_l, 'MIN':min, 'PTS':pts, 'FGM':fgm, 'FGA':fga, 'FG%':fgp, 
            '3PM':tpm, '3PA':tpa, '3P%':tpp, 'FTM':ftm, 'FTA':fta, 'FT%':ftp, 'OREB':oreb, 'DREB':dreb, 'REB':reb, 'AST':ast, 'STL':stl, 'BLK':blk,
            'TOV':tov, 'PF':pf, '+/-':pm},ignore_index=True)
        
```
