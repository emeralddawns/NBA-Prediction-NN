# Web Scrape NBA Game Data

```python
# Import dependencies
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
import time
```

```python
# Create a Web Driver Object
driver = webdriver.Chrome(ChromeDriverManager().install())
```

## Pull Regular Season Data

```python
# Connect to the website
driver.get(url="https://www.nba.com/stats/teams/boxscores/?Season=2021-22&SeasonType=Regular%20Season")
```

```python
# Select the drop down element that allows user to choose which page to view
table_section = driver.find_element(By.CLASS_NAME, "Crom_base__f0niE")
sel = Select(table_section.find_element(By.CLASS_NAME, "DropDown_select__4pIg9"))
```

```python
# Select the "ALL" option, which is Index 0
sel.select_by_visible_text("All")
# sel.select_by_index(0)
```

```python
# Read the game stats table on the webpage
time.sleep(1) # In case the table takes a while to load
stats = driver.find_element(By.CLASS_NAME, "Crom_table__p1iZz").get_attribute('outerHTML')
```

```python
# Convert the table (list) into a DataFrame
result = pd.read_html(stats)
```

```python
# Pull the table into a DataFrame
table_df = result[0]
table_df
```

```python
# Drop unnecessary columns
table_df = table_df.drop(['Match Up', 'Game Date', 'W/L', 'FG%', '3P%', 'FT%'], axis=1)
```

```python
# Rename the columns to match the database schema
table_df = table_df.rename(columns={"Team":"team"})
table_df = table_df.rename(columns={"MIN":"min", "PTS":"pts", "FGM":"fgm"})
table_df = table_df.rename(columns={"FGA":"fga", "3PM":"threepm", "3PA":"threepa", "FTM":"ftm", "FTA":"fta"})
table_df = table_df.rename(columns={"OREB":"oreb", "DREB":"dreb", "REB":"reb", "AST":"ast", "STL":"stl"})
table_df = table_df.rename(columns={"BLK":"blk", "TOV":"tov", "PF":"pf", "+/-":"point_diff"})
```

```python
# Export the dataframe into a CSV file
table_df.to_csv("Data/Scraped/NBA_Data_Reg_Season_2022.csv")
```
